# Partially Observable Markov Game

## Częściowo Obserwowalna Gra Markowa (POMG)

**Częściowo Obserwowalna Gra Markowa (POMG)** to matematyczne ramy do modelowania problemów sekwencyjnego podejmowania decyzji z udziałem wielu agentów, gdzie każdy agent ma niepełne informacje o środowisku. Rozszerza koncepcję Procesów Decyzyjnych Markowa (MDP) na przypadek wieloagentowy z częściową obserwowalnością. W POMG gra rozwija się poprzez stany, które nie są w pełni widoczne dla wszystkich agentów, a każdy agent podejmuje decyzje na podstawie własnych obserwacji, co prowadzi do strategicznych interakcji w warunkach niepewności.

Formalnie, POMG dla dwóch graczy definiuje się przez:
- **Stany ($S$)**: Zbiór możliwych stanów gry.
- **Akcje ($A_1$, $A_2$)**: Zbiory akcji dostępnych dla Gracza 1 i Gracza 2.
- **Funkcja Przejścia ($P$)**: Funkcja $P(s' | s, a_1, a_2)$, która podaje prawdopodobieństwo przejścia do stanu $s'$ ze stanu $s$, gdy Gracz 1 wykonuje akcję $a_1$, a Gracz 2 wykonuje akcję $a_2$.
- **Obserwacje ($O_1$, $O_2$)**: Zbiory możliwych obserwacji dla każdego gracza.
- **Funkcje Obserwacji ($O_1$, $O_2$)**: Funkcje takie jak $O_1(s)$, które mapują stan $s$ na obserwację dla Gracza 1 (podobnie dla Gracza 2), reprezentując częściowe informacje, które otrzymuje każdy gracz.
- **Funkcje Nagrody ($R_1$, $R_2$)**: Funkcje $R_1(s, a_1, a_2)$ i $R_2(s, a_1, a_2)$, które określają wypłaty dla każdego gracza w zależności od stanu i akcji.
- **Początkowy Rozkład Stanów**: Rozkład prawdopodobieństwa nad stanami startowymi.

W POMG agenci utrzymują historię swoich obserwacji i wybierają akcje na podstawie polityk, które mapują te historie na decyzje, często wymagając rozumowania o ukrytych informacjach i przewidywania zachowań innych.

## Poker Kuhn

[**Poker Kuhn**](https://en.wikipedia.org/wiki/Kuhn_poker) to minimalny, ale ilustracyjny wariant pokera, który exemplifies właściwości POMG. Zaangażowani są dwaj gracze i talia trzech kart (np. Król, Dama, Walet). Zasady są następujące:

1. **Rozdanie Kart**: Każdy gracz otrzymuje jedną kartę zakrytą, a trzecia karta pozostaje niewidoczna. Wprowadza to losowość na początku.
2. **Runda Licytacji**: Gracze na zmianę wykonują akcje—zazwyczaj **czekaj** lub **stawiaj** dla Gracza 1, oraz **pasuj**, **sprawdź**, lub **czekaj** dla Gracza 2, w zależności od wcześniejszych akcji—w jednej rundzie licytacji.
3. **Wynik**: Gra kończy się, gdy jeden gracz spasuje (drugi wygrywa pulę) lub po zakończeniu licytacji przez obu graczy, co prowadzi do **rozgrywki**, gdzie wyższa karta wygrywa pulę (np. Król > Dama > Walet).

Poker Kuhn to gra o **sumie zerowej**, co oznacza, że zysk jednego gracza jest stratą drugiego, a jego prostota czyni go idealnym polem do badania rozumowania strategicznego w warunkach niepewności.

## Pokera Kuhn a POMG

Poker Kuhn naturalnie pasuje do ram POMG w następujący sposób:

- **Stany ($S$)**: Stan można zdefiniować jako kombinację kart rozdanych obu graczom i sekwencję podjętych dotychczas akcji. Na przykład stan może być $(K, Q, \text{"Gracz 1 stawia"})$, gdzie Gracz 1 ma Króla, Gracz 2 ma Damę, a Gracz 1 postawił. Alternatywnie, stany mogą odpowiadać węzłom w drzewie gry, obejmującym zarówno rozdanie, jak i aktualną pozycję w sekwencji akcji.

- **Akcje ($A_1$, $A_2$)**: Dla Gracza 1 akcje obejmują **czekaj** lub **stawiaj** na początku. Dla Gracza 2 akcje zależą od wyboru Gracza 1 (np. **pasuj**, **sprawdź**, lub **czekaj**, jeśli Gracz 1 czeka; **pasuj** lub **sprawdź**, jeśli Gracz 1 stawia). Są to decyzje dostępne w każdym punkcie decyzyjnym.

- **Funkcja Przejścia ($P$)**: Przejścia są deterministyczne po rozdaniu kart. Na przykład, jeśli stan to $(K, Q, \emptyset)$ (początkowe rozdanie), a Gracz 1 stawia, następnym stanem staje się $(K, Q, \text{"Gracz 1 stawia"})$. Jedyna losowość występuje w początkowym rozdaniu kart (np. każde z 6 możliwych rozdań—$(K,Q)$, $(K,J)$, $(Q,K)$, itp.—ma prawdopodobieństwo $1/6$).

- **Obserwacje ($O_1$, $O_2$)**: Każdy gracz obserwuje swoją własną kartę i sekwencję podjętych dotychczas akcji, ale nie kartę przeciwnika. Na przykład, jeśli Gracz 1 ma Króla, a Gracz 2 stawia, obserwacja Gracza 1 to $(\text{"Mam Króla"}, \text{"Gracz 2 stawia"})$, podczas gdy obserwacja Gracza 2 to $(\text{"Mam Damę"}, \text{"Gracz 1 czeka, ja stawiam"})$. Ta częściowa informacja jest kluczowa dla niedoskonałej natury gry.

- **Funkcje Nagrody ($R_1$, $R_2$)**: Wypłaty zależą od końcowego stanu. Jeśli Gracz 2 pasuje po stawce Gracza 1, Gracz 1 wygrywa pulę (np. $+1$ dla Gracza 1, $-1$ dla Gracza 2). W rozgrywce wyższa karta wygrywa (np. jeśli Gracz 1 ma Króla, a Gracz 2 Damę, Gracz 1 dostaje $+2$, Gracz 2 dostaje $-2$, zakładając małą pulę i stawkę wielkości 1).

- **Początkowy Rozkład Stanów**: Gra zaczyna się z równomiernym rozkładem nad 6 możliwymi rozdaniami kart.

## Kluczowe Cechy POMG w Pokerze Kuhn

Poker Kuhn podkreśla podstawowe teoretyczne elementy POMG:

1. **Częściowa Obserwowalność**: Każdy gracz zna swoją kartę, ale nie kartę przeciwnika, co sprawia, że pełny stan (obie karty i historia akcji) jest tylko częściowo obserwowalny. To odzwierciedla funkcję obserwacji w POMG, gdzie gracze otrzymują niepełne informacje o stanie.

2. **Interakcja Wieloagentowa**: Dwaj gracze działają sekwencyjnie, a ich wybory wpływają na wyniki drugiego. Decyzja Gracza 1 o stawianiu lub czekaniu wpływa na opcje Gracza 2, co wymaga od obu przewidywania strategii drugiego.

3. **Sekwencyjne Podejmowanie Decyzji**: Gra rozwija się przez sekwencję akcji, a stan ewoluuje w zależności od tego, co robi każdy gracz. To zgadza się z własnością Markowa w POMG, gdzie następny stan zależy tylko od obecnego stanu i wspólnych akcji.

4. **Rozumowanie Strategiczne w Warunkach Niepewności**: Gracze muszą wnioskować o prawdopodobnej karcie przeciwnika na podstawie jego akcji. Na przykład, jeśli Gracz 2 sprawdzi stawkę, Gracz 1 może zaktualizować swoje przekonanie o sile karty Gracza 2. Optymalna gra wymaga **strategii mieszanych** (losowanie akcji), aby uniknąć przewidywalności, co jest charakterystyczne dla rozwiązań równowagi w POMG, takich jak równowagi Nasha.

```python
import numpy as np
from collections import defaultdict
from open_spiel.python import rl_environment
from open_spiel.python.rl_environment import StepType
import matplotlib.pyplot as plt
import sys

# Hyperparameters
initial_alpha = 0.5      # Initial Learning rate (higher for faster initial learning)
min_alpha = 0.01         # Minimum Learning rate
alpha_decay_episodes = 70000 # Episodes over which alpha decays
gamma = 1.0              # Discount factor (no discounting in Kuhn Poker)
initial_epsilon = 1.0    # Initial Exploration rate (start with full exploration)
min_epsilon = 0.05       # Minimum Exploration rate
epsilon_decay_episodes = 80000 # Episodes over which epsilon decays
num_episodes = 100000    # Number of training episodes
window_size = 1000       # Window size for moving average
print_every = 10000     # How often to print progress

# Epsilon-greedy action selection
def epsilon_greedy(Q_values, epsilon, num_actions):
    if np.random.rand() < epsilon:
        return np.random.randint(num_actions)
    best_action = np.argmax(Q_values)
    return best_action

game = "kuhn_poker"
env = rl_environment.Environment(game)
num_actions = env.action_spec()["num_actions"]

# Initialize Q-tables for both players
Q0 = defaultdict(lambda: np.zeros(num_actions))  # Player 0
Q1 = defaultdict(lambda: np.zeros(num_actions))  # Player 1

# Lists to store total rewards per episode
rewards_p0 = []
rewards_p1 = []

# Calculate decay rates per episode
alpha_decay_rate = (initial_alpha - min_alpha) / alpha_decay_episodes
epsilon_decay_rate = (initial_epsilon - min_epsilon) / epsilon_decay_episodes

print(f"Starting training for {num_episodes} episodes...")
for episode in range(num_episodes):
    # Calculate current alpha and epsilon
    alpha = max(min_alpha, initial_alpha - alpha_decay_rate * episode)
    epsilon = max(min_epsilon, initial_epsilon - epsilon_decay_rate * episode)

    if episode % print_every == 0:
         print(f"Episode {episode}/{num_episodes}, Alpha: {alpha:.4f}, Epsilon: {epsilon:.4f}")
         sys.stdout.flush()

    time_step = env.reset()
    actions_taken = [] # Keep track of (player, obs, action) for updates

    # Play one episode
    while not time_step.last():
        current_player = time_step.observations["current_player"]
        state = env.get_state # Use get_state for info state string
        obs = state.information_state_string(current_player)
        legal_actions = time_step.observations["legal_actions"][current_player] # Use legal actions if needed, although Kuhn always allows pass/bet

        Q = Q0 if current_player == 0 else Q1

        action = epsilon_greedy(Q[obs], epsilon, num_actions)
        actions_taken.append((current_player, obs, action)) # Store for potential later update if needed, but standard Q does immediate update

        # Environment step
        time_step = env.step([action])

        # --- Q-value Update ---
        # The update happens *after* the step, using the information *before* the step
        # We update the Q-value for the player who *just* acted
        prev_player, prev_obs, prev_action = actions_taken[-1]
        prev_Q = Q0 if prev_player == 0 else Q1
        reward = time_step.rewards[prev_player] # Reward received by the player who just acted

        if time_step.last():
            # Terminal state: The value of the next state is 0
            target = reward
        else:
            # Non-terminal state: Estimate value using the *next* player's Q-values
            next_player = time_step.observations["current_player"]
            next_state = env.get_state # State *after* the action
            next_obs = next_state.information_state_string(next_player)
            next_Q = Q0 if next_player == 0 else Q1

            # Ensure next observation is initialized in the next player's Q-table
            # next_Q[next_obs]

            # Use max Q-value from the *next* player's perspective for the next state
            max_next_q = np.max(next_Q[next_obs])
            target = reward + gamma * max_next_q

        # Apply the Q-learning update rule
        prev_Q[prev_obs][prev_action] += alpha * (target - prev_Q[prev_obs][prev_action])

    # End of episode: Store the final rewards
    rewards_p0.append(time_step.rewards[0])
    rewards_p1.append(time_step.rewards[1])

print(f"Episode {num_episodes}/{num_episodes}, Alpha: {alpha:.4f}, Epsilon: {epsilon:.4f}")
print("Training finished.")

# Compute moving average of rewards
def moving_average(data, window_size):
    if len(data) < window_size:
        print(f"Warning: Data length ({len(data)}) is less than window size ({window_size}). Adjusting window size.")
        window_size = len(data)
    if window_size == 0:
        return np.array([]) # Return empty if window_size becomes 0
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

ma_rewards_p0 = moving_average(rewards_p0, window_size)
ma_rewards_p1 = moving_average(rewards_p1, window_size)
ma_episodes = np.arange(window_size - 1, num_episodes)
plt.figure(figsize=(12, 6))
plt.plot(ma_episodes, ma_rewards_p0, label=f'Player 0 Moving Avg Reward (window={window_size})')
plt.plot(ma_episodes, ma_rewards_p1, label=f'Player 1 Moving Avg Reward (window={window_size})')

# Nash Equilibrium payoffs in Kuhn Poker are -1/18 for Player 0 and +1/18 for Player 1
nash_p0 = -1/18
nash_p1 = 1/18
plt.axhline(y=nash_p0, color='blue', linestyle='--', label=f'Nash Equilibrium P0 ({nash_p0:.4f})')
plt.axhline(y=nash_p1, color='orange', linestyle='--', label=f'Nash Equilibrium P1 ({nash_p1:.4f})')

plt.xlabel('Episode')
plt.ylabel(f'Average Reward (over {window_size} episodes)')
plt.title('Convergence of Independent Q-learning (with Decay) in Kuhn Poker')
plt.legend()
plt.grid(True)
plt.ylim(-0.5, 0.5) # Adjust ylim to better see convergence near equilibrium
plt.tight_layout()
plt.show()
```

### 5) Algorytm IQL dla pokera Kuhna przedstawiony w pseudokodzie uczy każdego gracza niezależnie, co może prowadzić do problemów, takich jak niestacjonarność środowiska. Jakie modyfikacje lub ulepszenia można wprowadzić do tego algorytmu, aby poprawić jego skuteczność w grze dwuosobowej, takiej jak poker Kuhna? Rozważ zarówno zmiany w strukturze uczenia, jak i potencjalne nowe techniki, np. modelowanie przeciwnika lub inne podejścia do uczenia ze wzmocnieniem.

- **Centralized Training with Decentralized Execution (CTDE):**  
  Podczas fazy treningowej umożliwienie agentom korzystania z globalnych informacji (np. zbioru doświadczeń obu agentów) w celu lepszej stabilizacji uczących się strategii, a następnie wykonanie działań lokalnie podczas gry.

- **Stosowanie stabilizujących mechanizmów, takich jak target network:**  
  Podobnie jak w algorytmach DQN, można wprowadzić powolno aktualizowaną sieć docelową, aby złagodzić skutki niestacjonarności środowiska.

- **Techniki podwójnego Q-learningu (Double Q-Learning):**  
  Mogą one zmniejszać nadmierną optymistyczną ocenę wartości akcji, co pomaga w dokładniejszym oszacowaniu strategii w środowisku, gdzie przeciwnik również się uczy.

- **Modelowanie przeciwnika (opponent modeling):**  
  Wprowadzenie mechanizmów pozwalających agentowi na eksplorację i oszacowanie strategii przeciwnika, aby móc dostosować swoją odpowiedź. Przykładem może być przyjmowanie założenia, że przeciwnik działa według określonego modelu, lub stosowanie technik iteracyjnych podobnych do doktryny „fictitious play”.

- **Algorytmy oparte o policy gradient:**  
  Metody te, zamiast szacować Q-wartości, bezpośrednio optymalizują parametry polityki. Są często bardziej odporne na problemy związane z niestacjonarnością, choć mogą wymagać innych sposobów eksploracji przestrzeni akcji.

- **Wspólne doświadczenie i replay buffer:**  
  Zastosowanie bufora doświadczeń pozwala na wielokrotne wykorzystanie danych treningowych, co może przyczynić się do lepszej stabilizacji aktualizacji.

Podsumowując, modyfikacje te mają na celu ograniczenie problemów wynikających z równoczesnej adaptacji obu agentów – poprawiając stabilność i zbieżność strategii do równowagi.
